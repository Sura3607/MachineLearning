```python
import imageio
import joblib            as jlb
import matplotlib.pyplot as plt
import numpy             as np
import pandas            as pd
import pydotplus         as pdp
import seaborn           as sbn

from IPython.display         import Image
from sklearn                 import tree
from sklearn.model_selection import train_test_split
from sklearn.tree            import DecisionTreeClassifier
```

```python
df = pd.read_excel(r'D:\Code_learning\MachineLearning\class\Ch03\Data\Iris.xls')
print(df.info())
df
```

```python
#visualize the data
colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'green', 'Iris-virginica': 'blue'}
plt.figure(figsize=(10, 6))
plt.scatter(df['petallength'], df['petalwidth'], c=df['iris'].map(colors))
for species in colors:
    plt.scatter([], [], c=colors[species], label=species)
plt.legend()
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris Dataset Classification')
plt.show()
```

```python
x = df[['sepallength', 'sepalwidth', 'petallength', 'petalwidth']]
y = df['iris']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
jlb.dump(model, 'DecisionTree_Iris.mdl')
```

```python
model = jlb.load('DecisionTree_Iris.mdl')
y_pred = model.predict(x_test)
print(y_pred)
y_test.head(10)
```
